```python
#Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
#scatter plot for age and fare columns in the titanic dataset.
```

```python
#Ans: 

import seaborn as sns
import plotly.express as px

# Load the titanic dataset using Seaborn
titanic_data = sns.load_dataset('titanic')

# Use Plotly Express to create a scatter plot for age and fare columns
fig = px.scatter(titanic_data, x='age', y='fare')

# Show the plot
fig.show()
```

```python
#Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
#Ans

import plotly.express as px
import plotly.graph_objs as go

# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a box plot using Plotly Express
fig = px.box(tips_data, x="day", y="total_bill", color="smoker")

# Show the plot
fig.show()
```

```python
#Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
#the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
#column with the color parameter.
```

```python
import plotly.express as px

# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a histogram using Plotly Express
fig = px.histogram(tips_data, x="sex", y="total_bill", color="day", pattern_shape="smoker")

# Show the plot
fig.show()
```

```python
#Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
```

```python
import plotly.express as px

# Load the iris dataset from Plotly
iris_data = px.data.iris()

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(iris_data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")

# Show the plot
fig.show()

```

```python
#Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python
#Ans: distplot is a function in the seaborn library that is used to plot a univariate distribution of observations. It combines a histogram of the data with a kernel density estimate, which is a non-parametric way to estimate the probability density function of a random variable.

#However, distplot has been deprecated in seaborn version 11.0.0 and replaced by displot, which is a more general function that can plot different types of distributions, including histograms, kernel density estimates, rug plots, and more.

#Here's an example of how to create a distplot using Plotly Express:
```

```python
import plotly.express as px

# Load the tips dataset from Plotly
tips_data = px.data.tips()

# Create a distplot using Plotly Express
fig = px.histogram(tips_data, x="total_bill", nbins=30, opacity=0.7, marginal="rug")

# Show the plot
fig.show()
```
