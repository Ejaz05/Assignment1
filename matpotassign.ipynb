```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#import numpy as np
#np.random.seed(3)
#x = 3 + np.random.normal(0, 2, 50)
#y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
#Ans : A scatter plot is a type of plot that displays the relationship between two variables as a collection of points on a two-dimensional plane. Each point in the scatter plot represents a single observation in the data, and the position of the point corresponds to the values of the two variables being plotted.

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

# generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# create scatter plot
plt.scatter(x, y)

# add title and axis labels
plt.title('Scatter Plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')

# display the plot
plt.show()

```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#Use the following data:
#import numpy as np
#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
#Ans: The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows us to display multiple plots side by side or in a grid layout, making it easier to compare different visualizations and analyze complex datasets.
import matplotlib.pyplot as plt
import numpy as np

# create data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create four subplots in a 2x2 grid layout
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# plot line 1
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Line 1')

# plot line 2
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Line 2')

# plot line 3
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Line 3')

# plot line 4
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Line 4')

# adjust spacing between subplots
fig.tight_layout()

# display the plot
plt.show()

```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#import numpy as np
#company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#profit = np.array([3000, 8000, 1000, 10000])
```

```python
#Ans: A bar plot, also known as a bar chart, is a visualization that represents categorical data with rectangular bars. The height or length of each bar corresponds to the value of the data being plotted, and the bars can be arranged horizontally or vertically.

#Bar plots are commonly used to compare different categories or groups of data, and to show trends or changes in data over time.

import matplotlib.pyplot as plt
import numpy as np

# create data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# create a horizontal bar plot
plt.barh(company, profit , color='red')
plt.title('Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

```python
#Ans: A box plot, also known as a box-and-whisker plot, is a type of plot that is used to display the distribution of data based on five key summary statistics: the minimum value, the first quartile (Q1), the median, the third quartile (Q3), and the maximum value. The box in the plot represents the interquartile range (IQR), which is the range between Q1 and Q3. The whiskers extend from the box to show the range of the data, excluding any outliers.

#Box plots are commonly used to compare the distribution of data across different groups or categories, and to identify any outliers or extreme values.

import matplotlib.pyplot as plt
import numpy as np

# create data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
