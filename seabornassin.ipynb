```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
#Ans: Scatter plot: A scatter plot is used to visualize the relationship between two numerical variables. It helps in identifying any correlation or trend between the variables. For example, we can use a scatter plot to see the relationship between the age and income of a population.

#Line plot: A line plot is used to visualize the trend of a numerical variable over time. It is useful for time series analysis and helps in identifying any seasonal or trend patterns. For example, we can use a line plot to visualize the daily stock prices of a company over a period of time.

#Bar plot: A bar plot is used to compare the frequency or distribution of a categorical variable. It is useful for showing the relative frequency of different categories. For example, we can use a bar plot to compare the number of students enrolled in different courses.

#Histogram: A histogram is used to visualize the distribution of a numerical variable. It is useful for identifying any skewness or outliers in the data. For example, we can use a histogram to visualize the distribution of the heights of a population.

#Heatmap: A heatmap is used to visualize the correlation between two variables. It is useful for identifying any patterns in the data. For example, we can use a heatmap to visualize the correlation between the performance of different students in a class.
```

```python
#Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
#Note: timepoint, signal, event, and region are columns in the fmri dataset.

#Ans:
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
#Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
#Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
#Ans:
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot the box plots
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)

```

```python
#Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
```

```python
#Ans:
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
```

```python
#Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
```

```python
#Ans:
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(data=iris, hue="species")
```

```python
#Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
#Ans: 
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
sns.heatmap(flights_matrix, annot=True, fmt="d")
```
