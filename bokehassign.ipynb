```python
#Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
#Ans:In Bokeh, glyphs are the basic visual building blocks that represent the data in a plot. They are visual shapes such as lines, circles, rectangles, and more. Glyphs can be used to represent different aspects of the data, such as x, y coordinates, size, color, and shape.

#To add glyphs to a Bokeh plot, you need to create a figure object and then call a glyph function on that object. The glyph function takes the data to be plotted as input and returns a glyph object that can be added to the figure.

#Here's an example of how to add a circle glyph to a Bokeh plot using Python code:

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 5, 4, 6, 3]

# Create a figure object
fig = figure(title="My Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add a circle glyph to the plot
fig.circle(x=x, y=y, size=10, color="navy", alpha=0.5)

# Show the plot
output_notebook()
show(fig)

```

```python
#Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

#Ans: 

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [2, 5, 4, 6, 3]

# Create a figure object
fig = figure(title="My Bokeh plot", x_axis_label="X-axis", y_axis_label="Y-axis")

# Customize the appearance of the axes
fig.xaxis.axis_label_text_font_size = "16pt"
fig.yaxis.axis_label_text_font_size = "16pt"
fig.xaxis.axis_line_color = "red"
fig.yaxis.axis_line_color = "blue"
fig.xaxis.major_tick_line_color = "red"
fig.yaxis.major_tick_line_color = "blue"
fig.xaxis.minor_tick_line_color = "red"
fig.yaxis.minor_tick_line_color = "blue"

# Customize the appearance of the title
fig.title.text_font_size = "20pt"
fig.title.text_font = "times"
fig.title.text_font_style = "italic"

# Customize the appearance of the legend
fig.circle(x=x, y=y, size=10, color="red", alpha=0.5, legend_label="Data")
fig.legend.label_text_font_size = "12pt"
fig.legend.location = "top_left"

# Show the plot
output_notebook()
show(fig)
```
