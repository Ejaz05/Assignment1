Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Ans: A t-test and a z-test are both statistical hypothesis tests used to determine the significance of differences between sample means and population means. However, they differ in the conditions under which they are appropriate.

A z-test is appropriate when the sample size is large (typically n > 30) and the population standard deviation is known. It assumes a normal distribution of data and uses the z-score formula to calculate the test statistic. A z-test is used to determine if the sample mean is significantly different from the population mean.

For example, suppose a manufacturing company produces a type of cable whose length is specified to be 100 cm. A sample of 50 cables is taken, and the length of each cable is measured. If the population standard deviation of cable lengths is known to be 3 cm, a z-test can be used to test if the sample mean length of cables is significantly different from 100 cm.

A t-test, on the other hand, is appropriate when the sample size is small (typically n < 30) or the population standard deviation is unknown. It assumes a normal distribution of data and uses the t-distribution to calculate the test statistic. A t-test is used to determine if the sample mean is significantly different from the population mean.

For example, suppose a researcher wants to test whether a new weight loss drug is effective in reducing body weight. The researcher recruits 25 participants and measures their body weight before and after taking the drug. Since the population standard deviation is unknown, a t-test can be used to test whether the mean change in body weight is significantly different from zero.

In summary, a z-test is appropriate when the sample size is large and the population standard deviation is known, while a t-test is appropriate when the sample size is small or the population standard deviation is unknown.





Q2: Differentiate between one-tailed and two-tailed tests.

Ans: A one-tailed test and a two-tailed test are both types of hypothesis tests used in statistics to determine whether a sample is significantly different from a population. The main difference between the two is in the directionality of the alternative hypothesis.

A one-tailed test, also called a directional test, is used when the alternative hypothesis is directional, meaning that it specifies the direction of the difference between the sample and the population. For example, if we want to test whether a new medication is more effective than an existing medication, the alternative hypothesis would state that the new medication is more effective than the existing medication. In this case, we would use a one-tailed test to determine whether the sample is significantly different from the population in the specified direction.

A two-tailed test, also called a non-directional test, is used when the alternative hypothesis is non-directional, meaning that it does not specify the direction of the difference between the sample and the population. For example, if we want to test whether a coin is fair, the alternative hypothesis would state that the coin is not fair, without specifying whether it is biased towards heads or tails. In this case, we would use a two-tailed test to determine whether the sample is significantly different from the population in either direction.

The choice between a one-tailed test and a two-tailed test depends on the research question and the nature of the alternative hypothesis. A one-tailed test has greater power to detect a significant difference in a specific direction, but it also has a greater risk of type I error (rejecting the null hypothesis when it is true) if the sample is significantly different from the population in the opposite direction. A two-tailed test is more conservative, but it can detect a significant difference in either direction.





Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Ans: In hypothesis testing, there are two types of errors that can occur: Type I error and Type II error.

Type I error, also known as a false positive, occurs when the null hypothesis is rejected when it is actually true. This means that the test concludes that there is a significant difference between the sample and the population, when in fact there is no such difference. The probability of making a Type I error is denoted by α (alpha), which is typically set at 0.05 or 0.01.

Example scenario: A pharmaceutical company tests a new drug for its effectiveness in treating a disease. The null hypothesis is that the drug is no more effective than a placebo, while the alternative hypothesis is that the drug is significantly more effective than a placebo. If the company incorrectly rejects the null hypothesis and concludes that the drug is effective, when in fact it is no more effective than a placebo, this would be a Type I error.

Type II error, also known as a false negative, occurs when the null hypothesis is not rejected when it is actually false. This means that the test fails to detect a significant difference between the sample and the population, when in fact there is such a difference. The probability of making a Type II error is denoted by β (beta).

Example scenario: A car manufacturer tests a new safety feature to determine whether it reduces the number of accidents. The null hypothesis is that the safety feature has no effect on the number of accidents, while the alternative hypothesis is that the safety feature significantly reduces the number of accidents. If the manufacturer fails to reject the null hypothesis and concludes that the safety feature has no effect, when in fact it does significantly reduce the number of accidents, this would be a Type II error.

In hypothesis testing, there is a trade-off between Type I and Type II errors. Reducing the probability of one type of error increases the probability of the other type of error. Therefore, it is important to choose an appropriate significance level and sample size to balance the risk of both types of errors.

Q4: Explain Bayes's theorem with an example.

Ans: Bayes's theorem is a mathematical formula used to calculate the probability of an event based on prior knowledge of conditions that might be related to the event. The theorem is named after Reverend Thomas Bayes, an 18th-century statistician.

Bayes's theorem can be stated as follows:

P(A|B) = P(B|A) x P(A) / P(B)

Where:

P(A|B) is the conditional probability of A given B
P(B|A) is the conditional probability of B given A
P(A) is the prior probability of A
P(B) is the prior probability of B
Let's consider an example to understand Bayes's theorem better. Suppose we have a medical test that can detect a certain disease. The test is not perfect and can give false positives (indicating the disease is present when it is not) and false negatives (indicating the disease is absent when it is actually present).

Suppose the prevalence of the disease in the population is 1%, and the test has a sensitivity of 95% and a specificity of 90%. This means that the test correctly identifies 95% of those who have the disease and 90% of those who do not have the disease.

Now, let's say that a patient tests positive for the disease. What is the probability that the patient actually has the disease?

Using Bayes's theorem, we can calculate this probability as follows:

Let A be the event "the patient has the disease"
Let B be the event "the patient tests positive for the disease"
We know that P(A) = 0.01 (the prevalence of the disease), and P(B|A) = 0.95 (the sensitivity of the test). We also know that P(B|not A) = 0.10 (the false positive rate), and P(not A) = 0.99 (the complement of the prevalence of the disease).

Using the formula, we can calculate:

P(A|B) = P(B|A) x P(A) / P(B|A) x P(A) + P(B|not A) x P(not A)
P(A|B) = 0.95 x 0.01 / (0.95 x 0.01 + 0.10 x 0.99)
P(A|B) = 0.086

Therefore, the probability that the patient actually has the disease given a positive test result is only 8.6%. This shows that even with a highly sensitive test, a positive result does not always mean that the patient has the disease.





Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

Ans: A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. In other words, it is a statistical measure that gives an estimated range of values that is likely to include an unknown population parameter, along with the probability that the true parameter falls within that range.

To calculate the confidence interval, the following information is required:

Sample mean (x̄)
Sample standard deviation (s)
Sample size (n)
Confidence level (CL)
The formula for calculating the confidence interval is:

CI = x̄ ± z*(s/√n)

Where:

CI: Confidence Interval
x̄: Sample mean
s: Sample standard deviation
n: Sample size
z: Z-score corresponding to the chosen confidence level (CL)
For example, let's say we want to calculate a 95% confidence interval for the average weight of a certain population of dogs, based on a sample of 50 dogs. The sample mean is 25 kg, and the sample standard deviation is 4 kg. The z-score corresponding to a 95% confidence level is 1.96.

Plugging in the values to the formula, we get:

CI = 25 ± 1.96*(4/√50)
CI = 25 ± 1.38
CI = (23.62, 26.38)

Therefore, we can be 95% confident that the true population mean weight of dogs falls between 23.62 kg and 26.38 kg.

t = (x̄ - μ) / (s / √n)

x̄ = 6 pounds
μ = 0 pounds
s = 2.5 pounds
n = 50 participants

t = (6 - 0) / (2.5 / √50) = 18.97

The degrees of freedom (df) for this test are (n - 1) = 49. Using a t-table or a t-distribution calculator, we can find the critical value for a one-tailed test with 49 degrees of freedom and a 95% confidence level to be 1.676.
