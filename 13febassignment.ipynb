```python
#Ans

#In Python, creating custom exceptions can help to handle specific types of errors or situations that may occur in a program. To create a custom exception, we need to define a new class that inherits from the base Exception class or one of its subclasses.

#The reason for using the Exception class as the base class for custom exceptions is that it provides a standard interface and behavior for all exceptions in Python. It defines common methods like __str__() and __repr__() that can be used to customize the error message and provide additional information about the exception.

#By using the Exception class as the base class, our custom exception inherits all the methods and attributes of the base class, and we can add additional functionality as needed. This makes it easier to write code that handles exceptions consistently and is more maintainable in the long run.
```

```python
#Q2. Write a python program to print Python Exception Hierarchy.
```

```python
def print_exception_hierarchy(base_exception=Exception, level=0):
    """Prints the exception hierarchy starting from the given base exception."""
    print('\t' * level + base_exception.__name__)
    for sub_exception in base_exception.__subclasses__():
        print_exception_hierarchy(sub_exception, level + 1)

# Print the hierarchy starting from the base Exception class
print_exception_hierarchy()
```

```python
#The LookupError class is used to define exceptions that are raised when a specified key or index is not found in a sequence or mapping. It is a base class for the KeyError and IndexError exceptions.

#KeyError: This error is raised when a key is not found in a dictionary. For example:
```

```python
try:
    
    my_dict = {'a': 1, 'b': 2, 'c': 3}
    value = my_dict['d']  # Raises KeyError
except KeyError as e:
    print(e)
```

```python
#IndexError: This error is raised when an index is out of range for a sequence (e.g., a list, tuple, or string). For example
```

```python
try:
    my_list = [1, 2, 3]
    value = my_list[3]  # Raises IndexError
except IndexError as e:
    print(e)
```

```python
#Ans:
#Here are some best practices for exception handling in Python:

#1 Catch exceptions at the appropriate level of granularity: Catch exceptions at the level where you can actually handle them. Avoid catching exceptions at a higher level than necessary, as this can make it harder to diagnose and debug errors.

#2 Use specific exception types: Use specific exception types (e.g., ValueError, TypeError, etc.) instead of catching the base Exception class. This allows you to handle different types of exceptions in different ways.

#3 Provide useful error messages: Provide informative error messages that help the user understand what went wrong and how to fix it.

#4 Use try-except blocks judiciously: Use try-except blocks only where necessary. Don't use them to handle expected conditions that can be checked with an if statement.

#5 Use the finally block for cleanup code: Use the finally block to execute cleanup code (e.g., closing files, releasing resources) regardless of whether an exception was raised or not.

#6 Don't catch exceptions silently: Avoid catching exceptions silently (i.e., without handling them). This can make it harder to diagnose and debug errors.

#7 Avoid using bare except: Avoid using bare except statements (i.e., except: with no exception type specified). This can catch unexpected exceptions and make it harder to diagnose and debug errors.

#8 Use context managers: Use context managers (e.g., with statements) to ensure that resources are properly acquired and released, even in the face of exceptions.

#9 Reraise exceptions when necessary: When catching exceptions, consider whether you should reraise the exception or raise a different exception. Reraising an exception can help to propagate the error up the call stack and provide more informative error messages.

#By following these best practices, you can write robust and maintainable code that handles exceptions gracefully and provides informative error messages.





```
